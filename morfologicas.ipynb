```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
src = cv.imread("j0.png")
src = cv.cvtColor(src, cv.COLOR_BGR2GRAY)
src2 = cv.imread("j1.png")
src2 = cv.cvtColor(src2, cv.COLOR_BGR2GRAY)
src3 = cv.imread("j2.png")
src3 = cv.cvtColor(src3, cv.COLOR_BGR2GRAY)
src4 = cv.imread("letras.png")
src4 = cv.cvtColor(src4,cv.COLOR_BGR2GRAY)
plt.imshow(src,cmap="gray")
```

```python
kernel = np.ones((5,5),np.uint8)
erosion = cv.erode(src,kernel,iterations=1)
fig,(ax1,ax2) = plt.subplots(1,2)
ax1.imshow(src,cmap="gray")
ax1.set_title("Org")
ax2.imshow(erosion,cmap="gray")
ax2.set_title("Erode")
```

```python
kernel = np.ones((5,5),np.uint8)
dilatacion = cv.dilate(src,kernel,iterations=1)
fig,(ax1,ax2) = plt.subplots(1,2)
ax1.imshow(src,cmap="gray")
ax1.set_title("Org")
ax2.imshow(dilatacion,cmap="gray")
ax2.set_title("Dilate")
```

```python
kernel = np.ones((5,5),np.uint8)
opening = cv.morphologyEx(src2, cv.MORPH_OPEN, kernel)

fig,(ax1,ax2) = plt.subplots(1,2)

ax1.imshow(src2,cmap="gray")
ax1.set_title("Org")
ax2.imshow(opening,cmap="gray")
ax2.set_title("Opening")
```

```python
kernel = np.ones((5,5),np.uint8)
close = cv.morphologyEx(src3, cv.MORPH_CLOSE, kernel)

fig,(ax1,ax2) = plt.subplots(1,2)

ax1.imshow(src3,cmap="gray")
ax1.set_title("Org")
ax2.imshow(close,cmap="gray")
ax2.set_title("Close")
```

```python
kernel = np.ones((5,5),np.uint8)
gradiente = cv.morphologyEx(src, cv.MORPH_GRADIENT, kernel)

fig,(ax1,ax2) = plt.subplots(1,2)

ax1.imshow(src,cmap="gray")
ax1.set_title("Org")
ax2.imshow(gradiente,cmap="gray")
ax2.set_title("Gradiente")
```

```python
distancia = cv.distanceTransform(src, cv.DIST_L2,0)

fig,(ax1,ax2) = plt.subplots(1,2)

ax1.imshow(src,cmap="gray")
ax1.set_title("Org")
ax2.imshow(distancia,cmap="gray")
ax2.set_title("Gradiente")
```

```python
kernel = np.ones((5,5),np.uint8)
src4_clean = cv.morphologyEx(src4, cv.MORPH_CLOSE, kernel, iterations=4) # limpiamos letras
src4_clean = cv.morphologyEx(src4_clean, cv.MORPH_OPEN, kernel, iterations=3) # limpiamos fondo

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.suptitle('')

ax1.imshow(src4, cmap='gray')
ax1.set_title("Org")
ax2.imshow(src4_clean, cmap='gray')
ax2.set_title("Opening")
```
