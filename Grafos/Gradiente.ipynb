```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calcula_y(x):
#y = x^2 + 7x -3
  return (x**2) + 7*x - 3

def derivada_y(x):
  return x*2 + 7
```

```python
nelem = 200
lx = np.linspace(-15, 10, nelem)
ly = calcula_y(lx)
plt.plot(lx, ly, color="r")
```

```python
x = -13
y = calcula_y(x)
print(f"x = {x} y = {y}")
lr = 0.25 #learning_rate
px = [x]
py = [y]
```

```python
x_incrementada = x + lr
y_incrementada = calcula_y(x_incrementada)
print(f"x_incrementada = {x_incrementada} y_incrementada = {y_incrementada}")
inc_x= x_incrementada - x
inc_y = y_incrementada - y
d = inc_y / inc_x
print(f"inc_x: {inc_x} inc_y: {inc_y} d: {d}")

x = x-lr * d
y = calcula_y(x)
print(f"inc_x: {x} inc_y: {y}")

px.append(x)
py.append(y)

plt.plot(lx, ly, color="r")
plt.scatter(px,py,color='b')
```
