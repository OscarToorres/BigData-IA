### Calculo de valor del salario correspodiente mediante la inflacion

```python
import numpy as np

pais_actual = "Alemania"
inflacion_acumulada = 0
salario_inflacion = []
indice_salario_base = 0

for index, pais in enumerate(df["pais"]):

    if pais == pais_actual:
        inflacion_acumulada = inflacion_acumulada +  (df["inflacion"].loc[index]/100)
        salario = df["salario"].loc[(df["pais"] == pais) & (df["fecha"] == 2000)][indice_salario_base].replace(" €", "").replace(".","")
        salario_relacion = np.int64(salario) * (1 + inflacion_acumulada)

        salario_inflacion.append(np.int64(salario_relacion))
    else:
        inflacion_acumulada = 0
        indice_salario_base = index
        inflacion_acumulada += (df["inflacion"].loc[index]/100)

        inflacion_acumulada = inflacion_acumulada +  (df["inflacion"].loc[index]/100)
        salario = df["salario"].loc[(df["pais"] == pais) & (df["fecha"] == 2000)][indice_salario_base].replace(" €", "").replace(".","")
        salario_relacion = np.int64(salario) * (1 + inflacion_acumulada)

        salario_inflacion.append(np.int64(salario_relacion))

    pais_actual = pais

```

```python
df["salario_inflacion"] = salario_inflacion
```
