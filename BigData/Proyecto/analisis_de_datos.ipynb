```python
nombres = []
df_paises = df.copy()
for i, name in enumerate(df["Country Name"]):
    if (name not in nombre_paises):
        df_paises = df_paises.drop(index=i)
    else:
        nombre_paises.remove(name)
```

```python
nombre_paises
```

```python
df_paises.shape
```

```python
df_paises = df_paises.dropna(axis=1)
df_paises.reset_index()
```

### Obtener para cada año, de cada pais, los valores correspondientes

```python
col_years = [1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]
valores = []

for index in df_paises.index:
    valor = []
    for col in col_years:
        valor.append(df_paises[str(col)][index])
    valores.append(valor)
```

```python
max = 0
for i in range(len(valores)):
    if (sum(valores[i]) > max):
        print(i)
        max = sum(valores[i])

print(max)
    
```
