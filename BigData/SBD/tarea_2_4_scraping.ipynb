## Obtencion de los datos de la tabla resultado

```python
def waiting():
    timeout = 30
    try:
        element_present = driver.find_element(By.XPATH,"//input[contains(@id,'viewns_Z7_AVEQAI930OBRD02JPMTPG21004_:form1:footerSiguiente')]")
        WebDriverWait(driver).until(element_present)
    except :
        print ("Timed out waiting for page to load")
        
```

```python
nombre_col = ["expediente","tipo de contrato", "estado", "importe", "fechas", "organo de contratacion"]
list_of_rows = []
table = True 
contador = 0
while table == True:
    tbody = driver.find_elements(By.XPATH,"//table[contains(@id,'myTablaBusquedaCustom')]/tbody/tr")
    if len(tbody) != 0:
        for i in range(len(tbody)):
            list_of_cells = []
            td = tbody[i].find_elements(By.TAG_NAME,"td")
            
            # Entrar a cada expediente y sacarle captura
            elem = driver.find_element(By.XPATH, '//a[2]/img')
            elem.click()
            WebDriverWait(driver, 10).until(lambda driver: len(driver.window_handles) == 2)
            driver.switch_to.window(driver.window_handles[1])
            wait = WebDriverWait(driver, 10)
            element = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/label")))
            contador += 1
            driver.save_screenshot(f'im/a{contador}.png')
            
            wait = WebDriverWait(driver, 10)
            element = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/label")))

            driver.close()
            driver.switch_to.window(driver.window_handles[0])

            for j in range(len(td)):
                list_of_cells.append(td[j].text)
            list_of_rows.append(list_of_cells)
        try:
            driver.find_element(By.XPATH,"//input[contains(@id,'viewns_Z7_AVEQAI930OBRD02JPMTPG21004_:form1:footerSiguiente')]").click()
        except:
            print("Fin de la tabla")
            table = False    
    else: break


dataFrame = pd.DataFrame(data = list_of_rows, columns = nombre_col)
dataFrame.to_csv("contrataciones.csv")
```

### Limpieza de datos

```python
from numpy import float64


for i in dataFrame.columns:
    dataFrame[i] = dataFrame[i].str.replace(r'(?<![A-Z "‚Äú()])([A-Z]+)', lambda x: f"\\n{x.group(1)}" if x.start() != 0 else x.group(1), regex=True)

dataFrame['importe'] = dataFrame['importe'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
```
