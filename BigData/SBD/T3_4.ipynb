### Calcular o endtime

```python
import datetime

endTime = []

for indice, fila in df.iterrows():

    timeParts = [int(s) for s in df['StartTime'][indice].split(':')]
    print(timeParts)

    totalMin = (timeParts[0] * 60 + timeParts[1])
    
    totalSecs, sec = divmod(totalMin, 60)
    print(totalSecs)
    print(sec)

    hr, min = divmod(totalSecs, 60)

    print("%d:%02d" % (hr, min))
```

5. Elimina as columnas na que máis da metade dos valores sexan nulos.

```python
mitad = len(df.columns)/2
#No borrará ninguno porque no hay ninguna fila con mas de 2 valores nulos
for i in df.index:
    if df.iloc[i].isna().sum() > mitad:
        df.drop(index=i)
```
