```python
def obtener_campos():

    lista_datos = []
    
    try:
        delito = WebDriverWait(driver, 0.5).until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class,'field field--node-field-crime field--name-field-crime field--type-entity-reference field--label-inline clearfix')]"))).text
    except:
        delito = ""
    try:
        nombre = WebDriverWait(driver, 0.5).until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class,'field field--node-field-w-first-name field--name-field-w-first-name field--type-string field--label-hidden field__item')]"))).text
    except:
        nombre = ""
    try:  
        pais = WebDriverWait(driver, 0.5).until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class,'field field--node-field-enfast-country field--name-field-enfast-country field--type-entity-reference field--label-hidden field__item')]"))).text
    except:
        pais = ""
    
    try:
        nacimiento = WebDriverWait(driver, 0.5).until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class,'field field--node-field-date-of-birth field--name-field-date-of-birth field--type-datetime field--label-inline clearfix')]"))).text
    except:
        nacimiento = ""

    try:    
        nacionalidad = WebDriverWait(driver, 0.5).until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class,'field field--node-field-nationality field--name-field-nationality field--type-entity-reference field--label-inline clearfix')]"))).text
    except:
        nacionalidad = ""

    try:    
        etnia = WebDriverWait(driver, 0.5).until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class,'field field--node-field-ethnic-origin field--name-field-ethnic-origin field--type-entity-reference field--label-inline clearfix')]"))).text
    except:
        etnia = ""
    try:
        idiomas = WebDriverWait(driver, 0.5).until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class,'field field--node-field-languages field--name-field-languages field--type-entity-reference field--label-inline clearfix')]"))).text
    except:
        idiomas = ""

    delito = delito.replace("DELITO Attempted ","")
    pais = pais.replace("by ","")
    nacimiento = nacimiento.replace("FECHA DE NACIMIENTO ","")
    nacionalidad = nacionalidad.replace("NACIONALIDAD ","")
    etnia = etnia.replace("ORIGEN ÉTNICO ", "")
    idiomas = idiomas.replace("IDIOMAS HABLADOS ","")

    lista_datos.append(delito)
    lista_datos.append(nombre)
    lista_datos.append(pais)
    lista_datos.append(nacimiento)
    lista_datos.append(nacionalidad)
    lista_datos.append(etnia)
    lista_datos.append(idiomas)

    return lista_datos
```

```python
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class, 'wanted_teaser_quick_info')]"))).click()

nombre_col = ["Nome_sospeitoso","país_buscando","data_nacemento", "nacionalidade", "etnia", "idiomas_hablados"]

list_of_rows = []

while True:

    list = driver.find_elements(By.XPATH,"//div[contains(@class, 'wanted_list')]/div")
        
    print(len(list))

    for i,data in enumerate(list):
        print(i)
        if (i > 1 and i < 5):
            data.click()
            # print("hago click")
            # action = WebDriverWait(data, 10).until(EC.presence_of_element_located((By.TAG_NAME,"div")))
            # print(action)
            # action.click()
            list_of_cells = obtener_campos()
            list_of_rows.append(list_of_cells)
        elif(i == 5):
            for i in range(3):
                data.click()
                # WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID,"downWanted"))).click()


```
