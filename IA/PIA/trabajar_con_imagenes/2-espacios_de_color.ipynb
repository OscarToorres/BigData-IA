```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
src1 = cv.imread('banda_vacia.png')
src1 = cv.cvtColor(src1,cv.COLOR_BGR2RGB)
src2 = cv.imread('banda_caja1.png')
src2 = cv.cvtColor(src2,cv.COLOR_BGR2RGB)
src3 = cv.imread('banda_caja1_5.png')
src3 = cv.cvtColor(src3,cv.COLOR_BGR2RGB)
src4 = cv.imread('banda_caja2.png')
src4 = cv.cvtColor(src4,cv.COLOR_BGR2RGB)

fig, (ax1,ax2,ax3,ax4) = plt.subplots(1,4)
ax1.imshow(src1)
ax2.imshow(src2)
ax3.imshow(src3)
ax4.imshow(src4)
```

```python
img1 = cv.cvtColor(src1,cv.COLOR_BGR2GRAY)
img2 = cv.cvtColor(src2,cv.COLOR_BGR2GRAY)

fig, (ax1,ax2,ax3) = plt.subplots(1,3)

resta_1 = img1 - img2
resta_2 = img2 - img1
resta_3 = cv.subtract(img2,img1)

ax1.imshow(resta_1,cmap="gray")
ax2.imshow(resta_2,cmap="gray")
ax3.imshow(resta_3,cmap="gray")
```

```python
img2 = cv.cvtColor(src2,cv.COLOR_BGR2GRAY)
img4 = cv.cvtColor(src4,cv.COLOR_BGR2GRAY)

fig, (ax1,ax2,ax3) = plt.subplots(1,3)

suma_1 = img2 + img4
suma_2 = img4 + img2
suma_3 = cv.subtract(img2,img4)

ax1.imshow(suma_1,cmap="gray")
ax2.imshow(suma_2,cmap="gray")
ax3.imshow(suma_3,cmap="gray")
```

```python
img1 = cv.cvtColor(src1, cv.COLOR_RGB2GRAY)
img2 = cv.cvtColor(src2, cv.COLOR_RGB2GRAY)
img4 = cv.cvtColor(src4, cv.COLOR_RGB2GRAY)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

r1 = cv.subtract(img2, img1)
r2 = cv.subtract(img4, img1)
r3 = cv.add(r1, r2)
#Para dar mas peso a una imagen mas que a otra
r3 = cv.addWeighted(r1, 0.5, r2, 0.5,gamma=0)


ax1.imshow(r1, cmap="gray")
ax2.imshow(r2, cmap="gray")
ax3.imshow(r3, cmap="gray")

```

```python
img1 = cv.cvtColor(src1, cv.COLOR_RGB2GRAY)
img2 = cv.cvtColor(src2, cv.COLOR_RGB2GRAY)
img3 = cv.cvtColor(src3, cv.COLOR_RGB2GRAY)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

r1 = cv.subtract(img2, img1)
r2 = cv.subtract(img3, img1)
r3 = cv.multiply(r1, r2)
#Para dar mas peso a una imagen mas que a otra
# r3 = cv.addWeighted(r1, 0.5, r2, 0.5,gamma=0)


ax1.imshow(r1, cmap="gray")
ax2.imshow(r2, cmap="gray")
ax3.imshow(r3, cmap="gray")
```

```python
img1 = cv.cvtColor(src1, cv.COLOR_RGB2GRAY)
img2 = cv.cvtColor(src2, cv.COLOR_RGB2GRAY)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

d1 = cv.divide(img2, img1)

ax1.imshow(img1, cmap="gray")
ax2.imshow(img2, cmap="gray")
ax3.imshow(d1, cmap="gray")
```

```python
src1 = cv.imread("test5.png"); src1 = cv.cvtColor(src1, cv.COLOR_BGR2RGB)
src2 = cv.imread("test6.png"); src2 = cv.cvtColor(src2, cv.COLOR_BGR2RGB)

img1 = cv.cvtColor(src1, cv.COLOR_BGR2GRAY)
img2 = cv.cvtColor(src2, cv.COLOR_BGR2GRAY)
img3 = cv.bitwise_or(img1, img2)
img4 = cv.bitwise_and(img1, img2)
img5 = cv.bitwise_xor(img1, img2)
img6 = cv.bitwise_not(img1)


fig , (ax1, ax2, ax3,ax4,ax5,ax6) = plt.subplots(1, 6)

ax1.imshow(img1, cmap="gray"); ax1.set_title("img1")
ax2.imshow(img2, cmap="gray"); ax2.set_title("img2")
ax3.imshow(img3, cmap="gray"); ax3.set_title("OR")
ax4.imshow(img4, cmap="gray"); ax4.set_title("AND")
ax5.imshow(img5, cmap="gray"); ax5.set_title("XOR")
ax6.imshow(img6, cmap="gray"); ax6.set_title("NOT")
```

```python
src1 = cv.imread("test.png"); src1 = cv.cvtColor(src1, cv.COLOR_BGR2RGB)
src2 = cv.imread("test5.png"); src2 = cv.cvtColor(src2, cv.COLOR_BGR2RGB)

img1 = cv.cvtColor(src1, cv.COLOR_BGR2GRAY)
img2 = cv.cvtColor(src2, cv.COLOR_BGR2GRAY)
img3 = cv.bitwise_and(img1, img2)
img4 = cv.multiply(img3, 1.5)
img5 = cv.bitwise_not(img2)
img5 = cv.bitwise_and(img1, img5)
img6 = cv.bitwise_or(img4, img5)

fig , (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(1, 6)

ax1.imshow(img1, cmap="gray"); ax1.set_title("img1")
ax2.imshow(img2, cmap="gray"); ax2.set_title("img2")
ax3.imshow(img3, cmap="gray"); ax3.set_title("img1 AND img2")
ax4.imshow(img4, cmap="gray"); ax4.set_title("X 1.5")
ax5.imshow(img5, cmap="gray"); ax5.set_title("Not img2 AND img1")
ax6.imshow(img6, cmap="gray"); ax6.set_title("img4 or ....")
```

```python
src1 = cv.imread("test.png"); src1 = cv.cvtColor(src1, cv.COLOR_BGR2RGB)
img1 = cv.cvtColor(src1, cv.COLOR_BGR2GRAY)
min, max, minLoc, maxLoc = cv.minMaxLoc(img1)
#umbral = (min + max)//2
#umbral = 127
media, desviacion = cv.meanStdDev(img1)
umbral = media


img1[img1 <= umbral] = 0
img1[img1 > umbral] = 255


plt.imshow(img1, cmap="gray")
```
