```python
import numpy as np
import matplotlib.pyplot as plt

#Matriz de correlacion
corr_matrix = df2.corr()

# Etiquetas
xlabs = corr_matrix.columns
ylabs = corr_matrix.columns
         
# Mapa de calor
fig, axes = plt.subplots(figsize=(25,20))
im = axes.imshow(corr_matrix)

# Agregar las etiquetas
axes.set_xticks(np.arange(len(xlabs)), labels = xlabs)
axes.set_yticks(np.arange(len(ylabs)), labels = ylabs)

# Agregar los valores a cada celda
for i in range(len(xlabs)):
    for j in range(len(ylabs)):
        v = round(corr_matrix.iloc[i, j], 2)
        axes.text(j, i, v,
                       ha = "center", va = "center", color = "w")
        
# Agregar la leyenda
cbar = axes.figure.colorbar(im, ax = axes)
# Rotar los nombres de las X
# axes.set_xticklabels(xlabs,rotation=90)
cbar.ax.set_ylabel("Valores", rotation = -90, va = "bottom")
```

### Entrenamos

```python
from sklearn.calibration import cross_val_predict
from sklearn.metrics import confusion_matrix

for clave, modelo in modelos.items():

    modelo = modelo.fit(x_train,y_train)
    y_predict = modelo.predict(x_test)
    # y_predict = cross_val_predict(modelo, x_train, y_train, cv=3)
    recall = recall_score(y_test,y_predict)
    print(f"{clave} Recall: {recall:.2f}")
    precission = precision_score(y_test, y_predict)
    print(f"{clave} Precission: {precission:.2f}") 

    mc = confusion_matrix(y_test,y_predict)
    print("=== Matriz de confusion ===")
    print(mc) 
```

```python
from sklearn.calibration import cross_val_predict
from sklearn.metrics import roc_curve
import seaborn as sns


for name, model in modelos.items():
    y_probas = cross_val_predict(model, x_train, y_train, cv=3,
                                    method="predict_proba")
    y_scores_forest = y_probas[:, 1]
    fpr, tpr, thresholds_forest = roc_curve(y_train, y_scores_forest)
    plt.plot(fpr, tpr, label=name)
    plt.legend()
```
